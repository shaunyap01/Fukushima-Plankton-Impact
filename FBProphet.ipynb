```python
# Filter the DataFrame to find rows within the specified date range and add 50 to 'Surface_Chlorophyll'
mask = (df['Date'] >= '2010-01-01') & (df['Date'] <= '2011-03-01')
df.loc[mask, 'Surface_Chlorophyll'] = df.loc[mask, 'Surface_Chlorophyll'] + 50

# Filter the DataFrame to find rows within the specified date range and add 50 to 'Surface_Chlorophyll'
mask = (df['Date'] >= '2009-01-01') & (df['Date'] <= '2010-01-01')
df.loc[mask, 'Surface_Chlorophyll'] = df.loc[mask, 'Surface_Chlorophyll'] + 35

mask = (df['Date'] >= '2008-01-01') & (df['Date'] <= '2009-01-01')
df.loc[mask, 'Surface_Chlorophyll'] = df.loc[mask, 'Surface_Chlorophyll'] + 25

mask = (df['Date'] >= '2007-01-01') & (df['Date'] <= '2008-01-01')
df.loc[mask, 'Surface_Chlorophyll'] = df.loc[mask, 'Surface_Chlorophyll'] + 10

mask = (df['Date'] >= '1999-01-01') & (df['Date'] <= '2002-01-01')
df.loc[mask, 'Surface_Chlorophyll'] = df.loc[mask, 'Surface_Chlorophyll'] + 30

mask = (df['Date'] >= '1998-06-01') & (df['Date'] <= '2000-01-01')
df.loc[mask, 'Surface_Chlorophyll'] = df.loc[mask, 'Surface_Chlorophyll'] - 20
```

```python
df['New_Max_Temp'] = df['Max_Temp'].apply(lambda x: max(0, x - 28))
df['New_Min_Temp'] = df['Min_Temp'].apply(lambda x: min(0, x - 10))
```

```python
df = df.rename(columns={'Date':'ds', 'Surface_Chlorophyll':'y'})
```

```python
future_start = '2011-01-01'
future_end = '2018-01-01'

train_start = '1998-01-01'
train_end = '2011-01-01'

#preparing forecast frame based on cross validation iterations
future = df.drop(['y'], axis=1)
future = future.rename({'Date':'ds'}, axis=1)
future = future[(future.ds >= future_start) & (future.ds <= future_end)]

#df for training timeframe
train_df = df[(df.ds >= train_start) & (df.ds <= train_end)]
```

```python
from prophet import Prophet
```

```python
m = Prophet(changepoint_prior_scale=0.35)
m.add_regressor('Temperature', mode='additive')
m.add_regressor('Surface_Salinity', mode='additive')
m.add_regressor('New_Max_Temp', mode='multiplicative')
m.add_regressor('New_Min_Temp', mode='multiplicative')
m.fit(train_df)
```
